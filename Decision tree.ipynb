```python
from sklearn.preprocessing import OneHotEncoder 
data = df

# Converting type of columns to category 
df['Area'] = df['Area'].astype('category') 
df['Gender'] = df['Gender'].astype('category')
df['Married'] = df['Married'].astype('category')
df['Education'] = df['Education'].astype('category')
df['Status'] = df['Status'].astype('category')
  
  
# Assigning numerical values and storing it in another columns 
df['Area_new'] = data['Area'].cat.codes 
data['Gender_new'] = data['Gender'].cat.codes 
data['Married_new'] = data['Married'].cat.codes 
data['Education_new'] = data['Education'].cat.codes 
data['Status_new'] = data['Status'].cat.codes
  
  
# Create an instance of One-hot-encoder 
enc = OneHotEncoder() 
  
# Passing encoded columns 
  
enc_data = pd.DataFrame(enc.fit_transform(data[['Gender_new', 'Area_new','Married_new','Education_new','Status_new']]).toarray()) 
  
# Merge with main 
New_df = df.join(enc_data) 
  
print(New_df) 
```
